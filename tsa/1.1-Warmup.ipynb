```python
import numpy as np
import matplotlib.pyplot as plt
```

### Exercise 1

```python
# 1.1. Generate 1000 samples from the standard normal
n = 1000
err = np.random.randn(n)
```

```python
# 1.2. Plot the result as a time series
plt.plot(err);
```

```python
# 1.3. Plot the result as a histogram
plt.hist(err, bins=20);
```

### Exercise 2

```python
# 2.1. Add a trendline to the noise
x = np.arange(n)
y = 0.01 * x + err
```

```python
# 2.2. Make a scatterplot
plt.scatter(x, y);
```

```python
# 2.3. Find and plot the best-fit line
f = np.polynomial.Polynomial.fit(x, y, 1, window=[0, n-1])
print(f)
plt.plot(f(x));
```

### Exercise 3

```python
# 3.1. Calculate the cumulative sum of the noise
walk = np.cumsum(err)
```

```python
#3.2. Plot the cumulative sum
plt.plot(walk)
```

###Exercise 4

```python
#4.1. Generate 1000 samples from a bivariate normal and make a scatterplot.
mu = np.zeros(2)
cov = np.array([[1,-0.5],[-0.5,2]])
err2 = np.random.multivariate_normal(mu, cov, 1000)
plt.scatter(err2[:,0], err2[:,1])
```

### Exercise 5

```python
#5.1. Calculate mean and covariance matrix of the bivariate normal samples.
mu_ = np.mean(err2, axis=0); print(f'mu_ = {mu_}')
cov_ = np.cov(err2.T); print(f'cov_ = \n{cov_}')
```
