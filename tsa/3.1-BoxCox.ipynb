```python
import numpy as np
import pandas as pd
from scipy.stats import boxcox
```

```python
df = pd.read_csv('airline_passengers.csv', index_col='Month', parse_dates=True)
print(df.head())
df.plot();
```

```python
df_sqrt = df.copy()
df_sqrt["Passengers"] = np.sqrt(df_sqrt["Passengers"])
df_sqrt.plot();
```

```python
df_log = df.copy()
df_log["Passengers"] = np.log(df_log["Passengers"])
df_log.plot();
```

```python
df_boxcox = df.copy()
df_boxcox["Passengers"], lmbda = boxcox(df_boxcox["Passengers"])
print(f'lmbda = {lmbda}')
df_boxcox.plot();
```

```python
df.hist(bins=20);
df_sqrt.hist(bins=20);
df_log.hist(bins=20);
df_boxcox.hist(bins=20);
```
