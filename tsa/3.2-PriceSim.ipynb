```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
T = 1000
P0 = 10
log_drift = 0.001
noise_sigma = 0.01
```

```python
# Simulate log prices
log_price = np.zeros(T+1)
log_price[0] = np.log(P0)
rng = np.random.default_rng()
for t in range(T):
  log_price[t+1] = log_price[t] + log_drift + noise_sigma * rng.standard_normal()

# Plot
plt.plot(np.exp(log_price));
plt.title("Simulated prices");
```
