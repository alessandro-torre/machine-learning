Note that we use transfer learning: the encoder is pre-trained on computer vision tasks (eg. classification) and its weights are frozen. We only need to train the decoder.

It makes sense to train encoder and decoder separately: we want the encoder to be good at feature extraction, and the decoder to do the actual segmentation job based on the features extracted by the encoder. The decoder can be properly trained to reproduce the segmentation masks only if it receives proper features from the encoder.
