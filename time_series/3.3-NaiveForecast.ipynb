```python
import numpy as np
import pandas as pd
from sklearn.metrics import mean_squared_error, \
  mean_absolute_error, r2_score, mean_absolute_percentage_error
```

```python
df = pd.read_csv('SPY.csv', index_col='Date', parse_dates=True)
df.head()
```

```python
# Naive forecast based on previous value
y_true = df["Close"].iloc[1:]
y_pred = df["Close"].shift(1).iloc[1:]
```

```python
# Metrics
print(f"MSE = {(y_pred - y_true).dot(y_pred - y_true) / len(y_true):.4f} = {mean_squared_error(y_pred, y_true):.4f}")
print(f"RMSE = {np.sqrt((y_pred - y_true).dot(y_pred - y_true) / len(y_true)):.4f} = {mean_squared_error(y_pred, y_true, squared=False):.4f}")
print(f"MAE = {(y_pred - y_true).abs().mean():.4f} = {mean_absolute_error(y_pred, y_true):.4f}")
print(f"R2 = {1 - mean_squared_error(y_pred, y_true) / y_true.var():.4f} = {r2_score(y_pred, y_true):.4f}")
print(f"MAPE = {np.mean((y_pred - y_true).abs() / y_pred.abs()):.6f} = {mean_absolute_percentage_error(y_pred, y_true):.6f}")
print(f"sMAPE = {np.mean((y_pred - y_true).abs() / (y_true.abs() + y_pred.abs()) * 2):.6f}")
```
